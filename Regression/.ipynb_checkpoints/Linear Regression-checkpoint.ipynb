```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import copy
import seaborn as sns
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from imblearn.over_sampling import RandomOverSampler
```

```python
df= pd.read_csv('SeoulBikeData.csv', encoding= 'unicode_escape').drop(['Date', 'Holiday', 'Seasons'], axis=1)
```

```python
df.head()
```

```python
dataset_cols = ["bike_count", "hour", "temp", "humidity", "wind", "visibility", "dew_pt_temp", "radiation", "rain", "snow", "functional"]
```

```python
df.columns = dataset_cols
df["functional"] = (df["functional"] == "Yes").astype(int)
df = df[df["hour"] == 12]
df = df.drop(["hour"], axis=1)
```

```python
df.head()
```

```python
df.head()
```

```python
for label in df.columns[1:]:
    plt.scatter(df[label], df['bike_count']) 
    plt.title(label)
    plt.ylabel('Bike Count at Noon')
    plt.xlabel(label)
    plt.show()
```

```python
df = df.drop(["wind", "visibility", "functional"], axis=1)
```

TRAINING/ VALIDATION/ TESTING

```python
train, validation, testing =np.split(df.sample(frac=1), [int(0.6*len(df)), int(0.8*len(df))])
```

```python
def get_x_y(dataframe, y_label, x_label=None):
    dataframe = copy.deepcopy(dataframe)
    if x_label is None:
        x = dataframe[[column for column in dataframe.columns if column != y_label]].values
    else:
        if len(x_label) == 1:
            x = dataframe[x_label].values.reshape(-1, 1)
        else:
            x = dataframe[x_label].values
    y = dataframe[y_label].values.reshape(-1, 1)
    data = np.hstack((x, y))
    return data, y, x
```

```python
_, y_train_temp, x_train_temp = get_x_y(train, 'bike_count', x_label=['temp'])
_, y_validation_temp, x_validation_temp = get_x_y(validation, 'bike_count', x_label=['temp'])
_, y_testing_temp, x_testing_temp = get_x_y(testing, 'bike_count', x_label=['temp'])
```

```python
temp_regression = LinearRegression()
temp_regression.fit(x_train_temp, y_train_temp)
```

```python
temp_regression.score(x_train_temp, y_train_temp)
```

```python
plt.scatter(x_train_temp, y_train_temp, label='Data', color='blue')
x = tf.linspace(-10, 50, 100)
plt.title('Bikes vs Temperature')
plt.plot(x, temp_regression.predict(np.array(x).reshape(-1,1)), label='fit', color='red', linewidth=3)
plt.xlabel('Temperature')
plt.ylabel('No. of Bikes')
plt.legend()
```

MULTIPLE LINEAR REGRESSION

```python
df.columns
```

```python
_, y_train_all, x_train_all = get_x_y(train, 'bike_count', x_label=df.columns[1:])
_, y_validation_all, x_validation_all = get_x_y(validation, 'bike_count', x_label=df.columns[1:])
_, y_testing_all, x_testing_all = get_x_y(testing, 'bike_count', x_label=df.columns[1:])
```

```python
all_regressor = LinearRegression()
all_regressor.fit(x_train_all, y_train_all)
```

```python
all_regressor.score(x_train_all, y_train_all)
```

LINEAR REGRSSION WITH NEURAL NET

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.xlabel('Epoch')
  plt.ylabel('MSE')
  plt.legend()
  plt.grid(True)
  plt.show()
```

```python
temp_normalizer = tf.keras.layers.Normalization(input_shape=(1,), axis = None)
temp_normalizer.adapt(x_train_temp.reshape(-1))
```

```python
temp_nn_model = tf.keras.Sequential([
    temp_normalizer, tf.keras.layers.Dense(1)])
```

```python
temp_nn_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1), loss='mean_squared_error')
```

```python
history = temp_nn_model.fit(x_train_temp.reshape(-1), y_train_temp, 
                            verbose=0, 
                            epochs=1000,
                            validation_data=(x_validation_temp, y_validation_temp))
```

```python
plot_loss(history=history)
```

```python
plt.scatter(x_train_temp, y_train_temp, label='Data', color='blue')
x = tf.linspace(-10, 50, 100)
plt.title('Bikes vs Temperature')
plt.plot(x, temp_nn_model.predict(np.array(x).reshape(-1,1)), label='fit', color='red', linewidth=3)
plt.xlabel('Temperature')
plt.ylabel('No. of Bikes')
plt.legend()
```

NEURAL NET

```python
temp_normalizer = tf.keras.layers.Normalization(input_shape=(1,), axis=None)
temp_normalizer.adapt(x_train_temp.reshape(-1))

nn_model = tf.keras.Sequential([
    temp_normalizer,
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])
nn_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
history = nn_model.fit(x_train_temp.reshape(-1), y_train_temp, 
                            verbose=0, 
                            epochs=100,
                            validation_data=(x_validation_temp, y_validation_temp))
```

```python
plot_loss(history)
```

```python
plt.scatter(x_train_temp, y_train_temp, label='Data', color='blue')
x = tf.linspace(-20, 50, 100)
plt.title('Bikes vs Temperature')
plt.plot(x, nn_model.predict(np.array(x).reshape(-1,1)), label='fit', color='red', linewidth=3)
plt.xlabel('Temperature')
plt.ylabel('No. of Bikes')
plt.legend()
```

```python
all_normalizer = tf.keras.layers.Normalization(input_shape=(6,), axis=-1)
all_normalizer.adapt(x_train_all)

nn_model = tf.keras.Sequential([
    all_normalizer,
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])
nn_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
history = nn_model.fit(x_train_all, y_train_all, 
                            verbose=0, 
                            epochs=100,
                            validation_data=(x_validation_all, y_validation_all))
```

```python
plot_loss(history)
```

```python
# calculate the MSE for both linear reg and nn
y_pred_lr = all_regressor.predict(x_testing_all)
y_pred_nn = nn_model.predict(x_testing_all)
```

```python
def MSE(y_pred, y_real):
  return (np.square(y_pred - y_real)).mean()
```

```python
MSE(y_pred_lr, y_testing_all)
```

```python
MSE(y_pred_nn, y_testing_all)
```
